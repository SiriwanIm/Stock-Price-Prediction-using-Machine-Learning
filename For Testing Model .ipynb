## Create X data (features)

Define features

```python
def create_feature(df,index_name):

  #1. Moving Average
  df['SME'] = df['Close'].rolling(window=10).mean()

  #2.RSI
  df['RSI'] = ta.RSI(np.array(df['Close']), timeperiod =14)

  #3. Stochastic Oscillator
  # Define periods
  k_period = 14
  d_period = 3
  # Adds a "n_high" column with max value of previous 14 periods
  df['n_high'] = df['High'].rolling(k_period).max()
  # Adds an "n_low" column with min value of previous 14 periods
  df['n_low'] = df['Low'].rolling(k_period).min()
  # Uses the min/max values to calculate the %k (as a percentage)
  df['%K'] = (df['Close'] - df['n_low']) * 100 / (df['n_high'] - df['n_low'])
  # Uses the %k to calculates a SMA over the past 3 values of %k
  df['%D'] = df['%K'].rolling(d_period).mean()
  df["Stochastic"] = df['%K'] - df['%D']
  df = df.drop(['n_high',"n_low"], axis=1)

  #4. Stochastic Correlation
  df['Sto_Corr'] = df['Close'].rolling(window=14).corr(df['Stochastic'])

  #5. Volatility
  df.set_index('Datetime', inplace=True)
  ret = 100 * (df.pct_change()[1:]['Close'])
  df['Volatility'] = ret.rolling(5).std()

  #6. MACD
  # Get the 26-day EMA of the closing price
  k = df['Close'].ewm(span=12, adjust=False, min_periods=12).mean()

  # Get the 12-day EMA of the closing price
  d = df['Close'].ewm(span=26, adjust=False, min_periods=26).mean()

  # Subtract the 26-day EMA from the 12-Day EMA to get the MACD
  macd = k - d

  # Get the 9-Day EMA of the MACD for the Trigger line
  macd_s = macd.ewm(span=9, adjust=False, min_periods=9).mean()

  # Calculate the difference between the MACD - Trigger for the Convergence/Divergence value
  macd_h = macd - macd_s

  # Add all of our new values for the MACD to the dataframe
  df['MACD_Con_Di'] = df.index.map(macd_h)

  #7. SAR
  df['SAR'] = ta.SAR(df.High.values, df.Low.values, acceleration = 0.02, maximum = 0.2)

  #8. Index_price
  if index_name == "set100":
      start = pd.to_datetime(start1) - pd.DateOffset(days=4)
      set100 = yf.download('^SET100.BK', start=start, end=end, interval='15m')
      set100.to_parquet('data.parquet')
      set100 = pd.read_parquet('data.parquet')
      set100= set100[["Open","High","Low","Close","Volume"]]
      df["Index_Price"] = set100['Close']
  else:
      start = pd.to_datetime(start1) - pd.DateOffset(days=4)
      set50 = yf.download('^SET50.BK', start=start, end=end, interval='15m')
      set50.to_parquet('data.parquet')
      set50 = pd.read_parquet('data.parquet')
      set50= set50[["Open","High","Low","Close","Volume"]]
      df["Index_Price"] = set50['Close']
  return df

```

## Prepare X data before using it to test the models
* Imputation data for missing value
* Drop NaN data

```python
from sklearn.impute import SimpleImputer

def prepare_data(df):

  #Fill infinity value in "Sto_Corr" Column
  df.replace(-np.inf, -1, inplace=True)
  df.replace(np.inf, 1, inplace=True)

  #Drop NaN data
  df.reset_index(drop=False, inplace=True)
  df = df.drop(range(33))

  #Imput data
  column_name = 'Index_Price'  # You can change this to the column you want to impute

  # Create a SimpleImputer instance with strategy='mean'
  imputer1 = SimpleImputer(strategy='mean')

  # Reshape the column to fit the imputer's input requirements
  column_to_impute = df[column_name].values.reshape(-1, 1)
  # Perform mean imputation
  df[column_name] = imputer1.fit_transform(column_to_impute)

  # Impute "Sto_Corr" data

  column_name = 'Sto_Corr'  # You can change this to the column you want to impute

  # Create a SimpleImputer instance with strategy='mean'
  imputer2 = SimpleImputer(strategy='mean')

  # Reshape the column to fit the imputer's input requirements
  column_to_impute = df[column_name].values.reshape(-1, 1)
  # Perform mean imputation
  df[column_name] = imputer2.fit_transform(column_to_impute)


  return df
```

## Test the model

* Load standard scaler and tranform X data
* Load model and use it to predict y
* Evaluate the model performance by comparing the predited y and true y data

```python
import joblib
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score, confusion_matrix,f1_score

def test_model(scaler_path,model_path,X_df,true_y_df):
  X_df.drop(['Datetime','index'], axis=1, inplace=True)

  #Stadardize the X data
  scaler = joblib.load(scaler_path)
  X_scaled= scaler.transform(X_df)

  #Predict with X data
  model = joblib.load(model_path)
  y_pred_proba = model.predict_proba(X_scaled)[:, 1]

  # Evaluate the Model
  auc_roc_score = roc_auc_score(true_y_df, y_pred_proba)

  y_pred = model.predict(X_scaled)

  # Calculate the confusion matrix
  conf_matrix = confusion_matrix(true_y_df,y_pred )

  # Calculate and print accuracy
  accuracy = accuracy_score(true_y_df, y_pred)

  #F-1 score
  f1_test = f1_score(true_y_df, y_pred)

  return print(f'AUC-ROC Score: {auc_roc_score}'), print("\nConfusion Matrix: ", conf_matrix),print("\nAccuracy:", accuracy),print('\nThe f1 score :', f1_test)

```
